```python
sub = df[['YearPublished', 'Complexity', 'AvgRating',
       'MinPlayers', 'MaxPlayers', 'NumOwned', 'MfgPlaytime',
       'MfgAgeRec', 'NumUserRatings', 'TotalMechanics',
       'boardgamerank', 'strategygamesrank', 'abstractsrank',
       'familygamesrank', 'thematicrank', 'cgsrank', 'wargamesrank',
       'partygamesrank', 'childrensgamesrank']]
```

```python
f,ax = plt.subplots(figsize=(10,10))
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```
